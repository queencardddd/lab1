```python
def check(s):
    stack = []

    for char in s:
        if char in "([{":
            stack.append(char)

        elif char in ")]}":
            if not stack:
                return False

            last = stack.pop()

            if (
                (char == ")" and last != "(") or
                (char == "]" and last != "[") or
                (char == "}" and last != "{")
            ):
                return False

    return len(stack) == 0


print(check("{[()]}"))   # True
print(check("{[(])}"))   # False
```
