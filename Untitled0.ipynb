```python
#Лабораторная работа 1.
#Александрова Екатерина ЦИБ-251
#1 Стек
#Реализовать проверку сбалансированности скобок (), {}, [] в строке.
#Используем встроенный тип list как стек.
#Методы: is_balanced(s: str) -> bool основная функция.
#Методы списка: append() - доюавление в стек, pop() - удаление из стека.
#Сложность операций - O(n).

def is_balanced(s: str) -> bool:
    stack = []
    pairs = {'(': ')', '{': '}', '[': ']'}

    for char in s:
        if char in pairs:
            stack.append(char)
        elif char in pairs.values():
            if not stack or pairs[stack.pop()] != char:
                return False

    return not stack
```
